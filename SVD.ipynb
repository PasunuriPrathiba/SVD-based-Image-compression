```
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as  np
import os
A = imread('/content/sample_data/Profile.jpg')

#converting RGB to GrayScale

X = np.mean(A,-1)
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```
U,S,VT = np.linalg.svd(X, full_matrices = False)
S = np.diag(S)


j=0
for r in (10,50,100,200,500) :
    
    #Performing low rank approximation
    
    X_app = U[:,:r] @ S[0:r, :r] @ VT[ :r, :]
    plt.figure(j+1)
    j+=1
    img = plt.imshow(X_app)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r=' +str(r))
    plt.show()
    
```

```
plt.figure(1)
plt.semilogy((np.diag(S)))
plt.title('singilar values')
plt.ylabel('log(\sigma_i)')
plt.xlabel('(i)')
plt.show()
```

From the figure it is clear that first few singular values are large in magnitude.
